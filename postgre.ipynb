```python
def get_records(birth_str: str, name: str):
    """
    QR에서 읽은 생년월일(birth_str: '19990409')과 이름(name: '채서린')으로
    medical_record 테이블에서 해당 환자의 모든 진료 기록을 조회.
    """
    sql = """
        SELECT patient_id, patient_name, diagnosis, visit_date, record_id
        FROM medical_record
        WHERE patient_id = %s
          AND patient_name = %s
        ORDER BY visit_date;
    """
    patient_id = int(birth_str)

    cur.execute(sql, (patient_id, name))
    rows = cur.fetchall()
    return rows

# 사용 예시
records = get_records("19990409", "채서린")
for r in records:
    print(r)

cur.close()
conn.close()
```
